Agrupar as questões por área de atuação do HumanizaSUS.
Para essa parte requer conhecimento de cada área e foi realizada em conjunto com a consultora do projeto.

```python
areas_humanizasus = {
    "Ambiência, Condições de Trabalho": [
        "Espaço físico",
        "Privacidade usuários",
        "Espaços comuns",
        "ambiente unidade",
        "condições de trabalho",
        "organização dos atendimentos"
    ],
    "Acolhimento, Valorização": [
        "relações interpessoais",
        "respeito ao usuário",
        "valorização saúde do trabalhador",
        "vínculo com pacientes",
        "acolhimento dos colegas",
        "gestão promove acolhimento",
        "relações de confiança",
        "momentos de escuta",
        "existem protocolos",
        "capacitação contínua"
    ],
    "Cogestão, Participação, Corresponsabilidade": [
        "participação nas decisões",
        "opinião ouvida",
        "responsabilidade compartilhada",
        "há corresponsabilidade"
    ],
    "Sustentabilidade, Cuidado Integral": [
        "práticas sustentáveis",
        "orientação descarte resíduos",
        "envolvimento práticas sustentáveis",
        "descarte resíduos perigosos",
        "preocupação saúde e meio ambiente"
    ]
}
```

```python
for area, perguntas in areas_humanizasus.items():
    df[area] = df[perguntas].mean(axis=1).round(2)

df.tail()
```

```python
nome_unidades = df["Unidade"].unique()
```

```python
unidade_media_area = []
for nome in nome_unidades:
    media1 = round(df[df["Unidade"] == nome]["Ambiência, Condições de Trabalho"].mean(), 2)
    media2 = round(df[df["Unidade"] == nome]["Acolhimento, Valorização"].mean(), 2)
    media3 = round(df[df["Unidade"] == nome]["Cogestão, Participação, Corresponsabilidade"].mean(), 2)
    media4 = round(df[df["Unidade"] == nome]["Sustentabilidade, Cuidado Integral"].mean(), 2)
    unidade_media_area.append({
        "Unidade": nome,
        "Ambiência, Condições de Trabalho": media1,
        "Acolhimento, Valorização": media2,
        "Cogestão, Participação, Corresponsabilidade": media3,
        "Sustentabilidade, Cuidado Integral": media4
    })
```

```python
df_geral_unidades = pd.DataFrame(unidade_media_area)
df_geral_unidades["Media Geral Unidade"] = df_geral_unidades[['Ambiência, Condições de Trabalho', 'Acolhimento, Valorização', 'Cogestão, Participação, Corresponsabilidade', 'Sustentabilidade, Cuidado Integral']].mean(axis=1).round(2).sort_values(ascending=True)
df_geral_unidades
```

```python
df_geral_unidades = df_geral_unidades.sort_values(by="Media Geral Unidade", ascending=False).reset_index(drop=True)
df_geral_unidades
```

```python
df_geral_unidades.to_excel("dados_finais/media_geral_unidade.xlsx")
```

```python
colunas_areas = [
    'Ambiência, Condições de Trabalho',
    'Acolhimento, Valorização',
    'Cogestão, Participação, Corresponsabilidade',
    'Sustentabilidade, Cuidado Integral'
]

desvio_por_area = df.groupby("Unidade")[colunas_areas].std().round(2)

desvio_por_area
```

```python
desvio_total_unidade = df.groupby("Unidade")[colunas_areas].apply(
    lambda x: x.values.flatten().std()
).round(2)
desvio_total_unidade
```

```python
df_desvios = desvio_por_area.copy()
df_desvios["Desvio Padrão Total Unidade"] = desvio_total_unidade
df_desvios = df_desvios.reset_index()
df_desvios = df_desvios.sort_values("Desvio Padrão Total Unidade")
df_desvios
```

```python
df_desvios.to_excel("dados_finais/desvio_padrao.xlsx", index=0)
```
